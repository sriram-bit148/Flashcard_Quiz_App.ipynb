```python
!pip install gradio -q

import gradio as gr

flashcards = [
    {"question": "What is HTML?", "answer": "HyperText Markup Language"},
    {"question": "What is CSS?", "answer": "Cascading Style Sheets"},
    {"question": "What is Python?", "answer": "A high-level programming language"}
]

current_index = 0

def get_question():
    return flashcards[current_index]["question"]

def show_answer():
    return flashcards[current_index]["answer"]

def next_card():
    global current_index
    current_index = (current_index + 1) % len(flashcards)
    return flashcards[current_index]["question"], ""

def previous_card():
    global current_index
    current_index = (current_index - 1) % len(flashcards)
    return flashcards[current_index]["question"], ""

def add_card(question, answer):
    flashcards.append({"question": question, "answer": answer})
    return "Card Added Successfully!"

def edit_card(question, answer):
    global current_index
    flashcards[current_index]["question"] = question
    flashcards[current_index]["answer"] = answer
    return "Card Updated Successfully!"

def delete_card():
    global current_index
    if len(flashcards) > 1:
        flashcards.pop(current_index)
        current_index = 0
        return "Card Deleted Successfully!"
    return "Cannot delete the last card."

with gr.Blocks() as app:
    gr.Markdown("# Flashcard Quiz App")

    question = gr.Textbox(label="Question", value=get_question())
    answer = gr.Textbox(label="Answer")

    with gr.Row():
        btn_answer = gr.Button("Show Answer")
        btn_prev = gr.Button("Previous")
        btn_next = gr.Button("Next")

    btn_answer.click(show_answer, outputs=answer)
    btn_prev.click(previous_card, outputs=[question, answer])
    btn_next.click(next_card, outputs=[question, answer])

    gr.Markdown("## Add / Edit Flashcards")

    new_question = gr.Textbox(label="Enter Question")
    new_answer = gr.Textbox(label="Enter Answer")

    with gr.Row():
        btn_add = gr.Button("Add Card")
        btn_edit = gr.Button("Edit Current Card")
        btn_delete = gr.Button("Delete Current Card")

    status = gr.Textbox(label="Status")

    btn_add.click(
        add_card,
        inputs=[new_question, new_answer],
        outputs=status
    )

    btn_edit.click(
        edit_card,
        inputs=[new_question, new_answer],
        outputs=status
    )

    btn_delete.click(
        delete_card,
        outputs=status
    )

app.launch()
```
